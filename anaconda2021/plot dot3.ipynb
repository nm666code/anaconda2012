```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y1= [48,46,48,45,46,46,44]
y2= [50,51,52,48,47,49,46]
y3= [53,55,56,49,48,50,48]
```

```python
plt.title('Weather')
plt.xlabel('Day')
plt.ylabel('temperature')

plt.plot(x,y1,label = 'y1')
plt.plot(x,y3,label = 'y3')
plt.plot(x,y2,label = 'y2')
plt.legend()
```
